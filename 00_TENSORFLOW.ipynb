### Shufle the order of the elements in a tensor


```python
# Shuffle a tensor
not_shuffled = tf.constant([[10,7], [3,4], [2,5]])

# Shuffle or non-shuffled
tf.random.shuffle(not_shuffled)
```

```python
not_shuffled
```

```python
random_3 = tf.random.Generator.from_seed(3)
random_3 = random_3.normal(shape=(2,3))
random_3
```

```python
tf.random.shuffle(random_3)
```

It looks like if we want our shuffled tensors to be in the same order, we've got to use the global level random seed as well as the operation level random seed

```python
tf.random.set_seed(42) # global level random seed
tf.random.shuffle(not_shuffled, seed = 42) #operation level random seed
```

### Indexing tensors
Tensors can be indexed just like Python lists

```python
# Get the first 2 elemnts from each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
# Alternative to tf.newaxis

tf.expand_dims(rank_4_tensor, axis = -1) # Last dim
```

```python
tf.expand_dims(rank_4_tensor, axis = 0) # first dim
```

### Matrix multiplication

There are two rules
* The inner dimensions must match
* The resulting matrix has the shape of the outter dimensions

```python
# Matrix Multiplication

tf.matmul(tensor, tensor)
```

```python
# Matrix Multiplication with Python operator "@"

tensor @ tensor
```

```python
x = tf.constant([[2,3], [3,4],[4,5]])
y = tf.constant([[2,3,4], [3,4,5]])
z = tf.constant([[2,3], [3,4],[4,5]])

x, y, z
```

```python
x @ y
```

```python
x @ z
```

```python
x @ tf.reshape(z, shape=(2,3))

#same as

tf.matmul(x, tf.reshape(z, (2,3) ))
```

```python
# Transpose

x, tf.transpose(x), tf.reshape(x,shape = (2,3))
```

# Forms of agregation
* Get the minimum
* Get the maximum
* Get the mean 
* Get the sum

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0,100, size = 50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# To find the variance of our tensor we can acces to tensorfow_probability
import tensorflow_probability as tfp
# Find the variance
tfp.stats.variance(E)

# Or 
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

```python
# Find the standar diviation
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

Find the positional maximum and minimum

```python
# Create a new tensor for finding positional m√≠nimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
F[tf.argmin(F)]
```
